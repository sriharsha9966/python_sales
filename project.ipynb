<center><big><h1>Sales Analysis Dashboard</h1></big></center>

<p>The dashboard displays key insights derived from sales data, providing a comprehensive view of customer demographics, product performance, and profitability. Through intuitive visualizations such as histograms, bar plots, and pie charts, it showcases the distribution of customer ages, variations in order quantities, the popularity of product categories, and the impact of customer age groups on purchasing behavior. Additionally, it highlights trends such as the monthly distribution of orders and regional customer distribution, enabling stakeholders to make informed decisions and optimize business strategies.</p>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sales = pd.read_csv('sales_data.csv')
```

<h1 style="color:darkgray">Distribution of Customer Ages</h1>

<p>Visualizes the spread of customer ages using a histogram, helping understand the age demographics of the customer base. Calculated using Pandas and plotted using Matplotlib.</p>

```python
plt.hist(sales['Customer_Age'],bins=5,edgecolor="black")
count,bins,_ = plt.hist(sales['Customer_Age'],bins=5)
for i in range(len(count)):
    bin_center = (bins[i]+bins[i+1])//2
    plt.text(bin_center,count[i],str(int(count[i])),ha='center',va='bottom')
plt.xlabel("Ages")
plt.ylabel("Values")
plt.title("Distribution of Customer Ages")
plt.show()
```

<h1 style="color:darkgray">Cusomer Age by Gender</h1>

<p>Depicts the distribution of customer ages for each gender using a boxplot, facilitating comparison between male and female customers. Implemented using Seaborn's boxplot function.</p>

```python
sns.boxplot(x='Customer_Gender', y='Customer_Age', data=sales)
plt.xlabel('Gender')
plt.ylabel('Customer Age')
plt.title('Variation of Customer Age by Gender')
plt.show()
```

<h1 style="color:darkgray">Month of Highest Number of Orders</h1>

<p>Illustrates the number of orders made in each month through a barplot, identifying the month with the highest order frequency. Achieved by counting occurrences of each month and plotting them using Matplotlib.</p>

```python
# Count number of orders per month
orders_per_month = sales['Month'].value_counts().sort_index()

# Plot barplot of number of orders per month
orders_per_month.plot(kind='bar', color='lightgreen')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.title('Number of Orders Per Month')
plt.show()
```

<h1 style="color:darkgray">Distribution of Order Quantitites</h1>

<p>Represents the frequency distribution of order quantities using a histogram, revealing common order sizes. Plotted using Matplotlib after extracting order quantity data.</p>

```python
# Plot histogram of order quantities
plt.hist(sales['Order_Quantity'], bins=20, color='salmon', edgecolor='black')
plt.xlabel('Order Quantity')
plt.ylabel('Frequency')
plt.title('Distribution of Order Quantities')
plt.show()
```

<h1 style="color:darkgray">Most Common Product Categories</h1>

<p>Displays the proportion of each product category in the dataset using a pie chart, highlighting the most popular product categories. Computed the frequency of each category and plotted it with Matplotlib.</p>

```python
# Count number of occurrences of each product category
product_category_counts = sales['Product_Category'].value_counts()

# Plot pie chart of product categories
plt.pie(product_category_counts, labels=product_category_counts.index, autopct='%1.1f%%')
plt.title('Most Common Product Categories')
plt.axis('equal')
plt.show()
```

<h1 style="color:darkgray">Profit Variation Across Product Categories</h1>

<p>Shows the average profit for each product category using a barplot, identifying which categories generate higher profits. Calculated mean profit for each category and plotted it with Matplotlib.</p>

```python
# Group data by product category and calculate mean profit
profit_by_category = sales.groupby('Product_Category')['Profit'].mean()

# Plot barplot of mean profit by product category
profit_by_category.plot(kind='bar', color='lightblue')
plt.xlabel('Product Category')
plt.ylabel('Mean Profit')
plt.title('Profit Variation Across Product Categories')
plt.show()
```

<h1 style="color:darkgray">Distribution of Unit Costs</h1>

<p>Visualizes the distribution of unit costs using a histogram, providing insight into the cost structure of products. Created using Matplotlib after extracting unit cost data.</p>

```python
# Plot histogram of unit costs
plt.hist(sales['Unit_Cost'], bins=20, color='orange', edgecolor='black')
plt.xlabel('Unit Cost')
plt.ylabel('Frequency')
plt.title('Distribution of Unit Costs')
plt.show()
```

<h1 style="color:darkgray">Country with the Highest Number of Customers</h1>

<p>Displays the number of customers in each country using a barplot, highlighting the country with the most customers. Calculated customer counts for each country and plotted them with Matplotlib.</p>

```python
# Count number of customers per country
customers_per_country = sales['Country'].value_counts().sort_values(ascending=False)

# Plot barplot of number of customers per country (top 10)
customers_per_country.head(10).plot(kind='bar', color='lightcoral')
plt.xlabel('Country')
plt.ylabel('Number of Customers')
plt.title('Country with the Highest Number of Customers')
plt.show()
```

<h1 style="color:darkgray">Distribution of Profits </h1>

<p>Illustrates the distribution of profits using a histogram, showcasing the spread of profit margins. Plotted using Matplotlib after extracting profit data.</p>

```python
# Plot histogram of profits
plt.hist(sales['Profit'], bins=20, color='purple', edgecolor='black')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.title('Distribution of Profits')
plt.show()
```

<h1 style="color:darkgray">Impact of Customer Age Group on Product Category </h1>

<p>Demonstrates the relationship between customer age groups and product categories using a stacked barplot, indicating which categories are preferred by different age groups. Constructed a pivot table and plotted it as a stacked barplot using Pandas and Matplotlib.</p>

```python
# Create a pivot table of counts of product categories by age group
pivot_table = pd.pivot_table(sales, index='Age_Group', columns='Product_Category', aggfunc='size', fill_value=0)

# Plot stacked barplot
pivot_table.plot(kind='bar', stacked=True)
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Impact of Customer Age Group on Product Category')
plt.legend(title='Product Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
